### Toss Analysis  
- How many times did the toss winner also win the match?
- When teams chose to bat vs field — which decision led to more wins?
- Which team won the most tosses in 2022?



```python
# Toss win == match win

df['toss_winner == match_winner'] = (df['toss_winner'] == df['match_winner']).map({
    True: 'Yes',
    False: 'No'
})


toss_win_match_win = df[df['toss_winner'] == df['match_winner']]
total = toss_win_match_win['match_winner'].value_counts()
total_valid_matches = df.dropna(subset=['toss_winner', 'match_winner']).shape[0]
toss_impact = (toss_win_match_win.shape[0] / total_valid_matches) * 100

print(f"{toss_win_match_win.shape[0]} times, the toss winner also win the matches")
print("\nTeams that win the toss and also matches")
print(total)
print(f"Total = {len(toss_win_match_win)}")

print(f"Total valid matches: {total_valid_matches}")
print(f"Toss impact percentage: {toss_impact:.2f}%%")


#visualization
plt.figure(figsize=(14,6))
ax = sns.barplot(x=total.index, y=total.values, palette='Set2', edgecolor='black')
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)
plt.title("Toss Win == Match Win", fontsize=18, fontweight='bold', family='DejaVu Sans')
plt.xlabel("Match Winner Teams")
plt.ylabel("Count")
plt.tight_layout()
```

```python
plt.pie(total.values,autopct='%1.1f%%', labels = total.index, explode=[0.07,0,0,0,0,0,0,0,0,0], shadow=True)
```

```python
values = [toss_impact, (100 - toss_impact)]
labels = ['Toss Winner = Match Winner', 'Toss Winner ≠ Match Winner']
colors = ['#ff6666', '#66b366']  # softer red & green

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=colors,
        startangle=90, explode=(0.04, 0), shadow=True)

plt.title('Impact of Toss on Match Outcome', fontsize=14, fontweight='bold')
plt.show()
```

```python
# When teams chose to bat vs field — which decision led to more wins?
decision = df[df['toss_winner == match_winner'] == 'Yes']['toss_decision'].value_counts()

print(f"Toss Decision where Toss winner == Match winner")
print(decision)
dec_pct = (decision / decision.sum()) * 100
for dec, pct in dec_pct.items():
    print(f"{dec}: {pct:.2f}%%")


fig, axes = plt.subplots(1,2,figsize=(14,6))
fig.suptitle("which toss decision won more matches?", fontweight='bold', fontsize='18')

my_ax = sns.barplot(x=decision.index, y=decision.values, palette='hsv', ax=axes[0], edgecolor='black')

for container in my_ax.containers:
    my_ax.bar_label(container, fmt='%d', padding=3)

axes[0].set_title("BarChart", fontweight='bold', fontsize=12)
axes[0].set_xlabel("toss_decision(toss winner == match winner)")
axes[0].set_ylabel("Count")


axes[1].pie(decision.values,labels=decision.index, autopct='%1.0f%%', colors=['#ff6666', '#66b366'], explode=[0,0.04], shadow=True)
axes[1].set_title("PieChart", fontweight='bold', fontsize=12)


plt.tight_layout()
plt.legend()
plt.show()
    
```

```python
#Which team won the most tosses in 2022?
tosses_won = df['toss_winner'].value_counts()
print(f"Teams that won the most tosses in 2022 IPL are: ")
print(tosses_won.head(2))

fig,axes = plt.subplots(1,2,figsize=(16,6))
fig.suptitle("Number of Tosses won by teams", fontsize=20, fontweight='bold')

ax = sns.barplot(x=tosses_won.index, y=tosses_won.values, palette='hot', edgecolor='black', ax=axes[0])
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)

axes[0].tick_params(axis='x', rotation=45)
axes[0].set_title("BarChart", fontsize=15,fontweight='bold')
axes[0].set_xlabel("Teams")
axes[0].set_ylabel('Count')

axes[1].pie(tosses_won.values,labels=tosses_won.index, autopct='%1.2f%%', explode=[0.07,0.04,0,0,0,0,0,0,0,0], shadow=True, startangle=140)
axes[1].set_title("PieChart", fontsize=15,fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#Toss Decision
toss_decision = df['toss_decision'].value_counts()
print(toss_decision)
decision_pct = (toss_decision / toss_decision.sum()) * 100
print(f"\nPercentage breakdown:")
for decision, pct in decision_pct.items():
    print(f"{decision}: {pct:.2f}%")


fig,axes = plt.subplots(1,2,figsize=(14,6))
fig.suptitle("Toss Decision (Bat vs Field)", fontsize=20, fontweight='bold')

ax = sns.barplot(x=toss_decision.index, y=toss_decision.values, palette='hot', edgecolor='black', ax=axes[0])
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)

axes[0].set_title("BarChart", fontsize=15,fontweight='bold')
axes[0].set_xlabel("Decision")
axes[0].set_ylabel('Count')

axes[1].pie(toss_decision.values,labels=toss_decision.index, autopct='%1.2f%%', explode=[0.07,0], shadow=True)
axes[1].set_title("PieChart", fontsize=15,fontweight='bold')

plt.tight_layout()
plt.legend()
plt.show()
```

### Team Performance 
- Which team won the most matches in 2022?
- Which team lost the most?
- Win Percentage for each team?

```python
#won most matches in 2022
match_won = df['match_winner'].value_counts()

print("Most Matches Won by teams")
print(match_won)

plt.figure(figsize=(12,6))
ax = sns.barplot(x=match_won.index, y=match_won.values,palette='hsv', edgecolor='black')
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)

plt.title("Matches won by teams", fontsize=18, fontweight ='bold')
plt.xlabel("Teams")
plt.ylabel("Count")
plt.tick_params(axis='x', rotation=45)
```

```python
# Total matches Played by each teams 

matches_played = pd.concat([df['team1'], df['team2']]).value_counts()

plt.figure(figsize=(12,6))
ax = sns.barplot(x=matches_played.index, y=matches_played.values, palette='gist_ncar', edgecolor='black')
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)

plt.title("Total Matches Played by each teams", fontsize=18, fontweight='bold')
plt.xlabel("Teams")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
#Match Summary Data Frame

matches_won = df['match_winner'].value_counts()
matches_lost = matches_played - matches_won
win_pct = ((match_won / matches_played) * 100).round(2)
lost_pct = 100 - win_pct

summary = pd.DataFrame({
    'Matches Played' : matches_played, 
    'Matches Won' : match_won,
    'Matches Lost' : matches_lost,
    'Win %' : win_pct, 
    'Lost %' : lost_pct
}).fillna(0)

summary.reset_index(inplace=True)

summary.rename(columns={
    'index' : 'Teams'
}, inplace=True)

summary
```

```python
#Melting Summary DF for 3bars BarChart 
summary_long = summary.melt(
    id_vars='Teams',
    value_vars=['Matches Played','Matches Won','Matches Lost'],
    var_name='Outcome',
    value_name='Count'
)

summary_long

```

```python
#Plot Grouped bar Charts
plt.figure(figsize=(16,7))
ax = sns.barplot(x='Teams', y='Count', hue='Outcome', data=summary_long, palette='hot')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)

ax.set_title('Matches Played vs Won vs Lost per Team', fontsize=15, fontweight='bold')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Which team lost the most?
print(f"Teams Lost the most Matches")
print(summary.sort_values(by='Matches Lost', ascending=False, ignore_index=True)[['Teams','Matches Lost']].head(5))
print()

plt.figure(figsize=(12,6))
ax = sns.barplot(data=summary,x='Teams',y='Matches Lost', palette='gist_ncar')
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)

plt.title("Matches Lost By Teams")
plt.legend()
plt.show()
```

```python
#Win Percentage for each team? 
print(f"Win percentage for each team")
print(summary[['Teams','Win %']])
print()


plt.figure(figsize=(12,6))
ax = sns.barplot(data=summary, x='Teams', y='Win %', palette='hot_r')
for container in ax.containers: 
    ax.bar_label(container, fmt='%.2f', padding=4)

plt.title("Win % of each team", fontsize=14, fontweight='bold')
plt.show()
```

### Match Results Analysis
- How many matches were won batting first vs chasing?
- What was the biggest winning margin (runs)?
- What was the biggest winning margin (wickets)?


```python
#How many matches were won batting first vs chasing? 
won = df['won_by'].value_counts()
print("Matches won by batting first vs chasing")
print(won)

ax = sns.barplot(x=won.index, y=won.values, palette='hsv')
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)

plt.title("Wickets vs Runs")
plt.ylabel("Count")
```

```python
#What was the biggest winning margin (runs)? 
won_by_runs = df[df['won_by'] == 'Runs'].sort_values(by='margin', ascending=False).head()[['match_winner','margin', 'won_by']]
print("Top 5 biggest win by runs")
print(won_by_runs)


plt.figure(figsize=(12,6))
ax = sns.barplot(data=won_by_runs, x='match_winner', y='margin', ci=None, palette='hsv')
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)

plt.title("Top 5 team with biggest margin wins by runs", fontweight = 'bold', fontsize=14)
```

```python
# What was the biggest winning margin (wickets)?
won_by_wickets = df[df['won_by'] == 'Wickets'].sort_values(by='margin', ascending=False).head()[['match_winner', 'margin', 'won_by']]
print("Top 5 teams with biggest margin wins by wickets")
print(won_by_wickets)
print()

plt.figure(figsize=(12,6))
ax = sns.countplot(won_by_wickets['match_winner'])
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=4)

plt.title("Top 5 team with biggest margin wins by wickets", fontweight = 'bold', fontsize=14)
```

### Innings Score Analysis 
- What was the average first innings score across all matches?
- What is the highest and lowest first innings score?
- Is there a pattern — high first innings score = harder to chase?

```python
#What was the average first innings score across all matches?
avg_first_ing_score = df['first_ings_score'].mean()
max_first_ing_score = df['first_ings_score'].max()
min_first_ing_score = df['first_ings_score'].min()


print(f"Average First Inning Score: {avg_first_ing_score:.2f}")
print(f"Maximum First Inning Score: {max_first_ing_score}")
print(f"Minimum First Inning Score: {min_first_ing_score}")
print()


plt.figure(figsize=(12,6))
sns.histplot(data=df, x='first_ings_score', bins=30, color='purple', alpha=0.9, edgecolor='black', kde=True)
plt.axvline(avg_first_ing_score, color='red', linewidth=2, linestyle='dashed', label=f"Mean: {avg_first_ing_score:.2f}")
plt.axvline(max_first_ing_score, color='green', linewidth=2, linestyle='dotted', label=f"Max: {max_first_ing_score}")
plt.axvline(min_first_ing_score, color='orange', linewidth=2, linestyle='solid', label=f"Min: {min_first_ing_score}")

plt.title("Distribution of First Innings Scores", fontweight='bold', fontsize=14)
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Which venue produces the highest average first innings score?
venue_first_avg = df.groupby('venue')['first_ings_score'].mean().sort_values(ascending=False)
print("Highest average first innings score by venue")
print(venue_first_avg)
print()

plt.figure(figsize=(12,6))
ax = sns.barplot(x=venue_first_avg.index, y=venue_first_avg.values, palette='hsv', edgecolor='black')
for container in ax.containers: 
    ax.bar_label(container, fmt='%.2f', padding=4)

plt.tick_params(axis='x', rotation=60)
plt.title("Highest average first innings score by venue", fontsize=12, fontweight='bold')
plt.ylabel("Score")
plt.show()
```

```python
#Score distribution — first innings vs second innings - Histplot

plt.figure(figsize=(14,6))
sns.histplot(df['first_ings_score'], bins = 20, linewidth=2, edgecolor='black', color='red', kde=True, alpha=0.5, label='First Innings Score')
sns.histplot(df['second_ings_score'], bins = 20, linewidth=2, edgecolor='black', color='orange', kde=True, alpha=0.5, label='Second Innings Score')

plt.title("first_ings_score vs second_ings_score")
plt.xlabel("Score")
plt.legend()
plt.show()
```

```python
#Score distribution — first innings vs second innings - Boxplot 


# Reshape into long format
score_df = pd.DataFrame({
    'First Innings': df['first_ings_score'],
    'Second Innings': df['second_ings_score']
})
score_long = score_df.melt(var_name='Innings', value_name='Score')

# Plot boxplot
plt.figure(figsize=(10,6))
ax = sns.boxplot(data=score_long, x='Innings', y='Score', palette=['red','orange'])
plt.title("Score Distribution — First vs Second Innings")

# Calculate and annotate 5-number summary for each category
for i, inning in enumerate(score_df.columns):
    values = score_df[inning].dropna()
    summary = np.percentile(values, [0,25,50,75,100])  # min, Q1, median, Q3, max
    
    # Annotate each value on the plot
    for val, label in zip(summary, ['Min','Q1','Median','Q3','Max']):
        ax.text(i, val, f"{label}: {val}", 
                ha='center', va='bottom', fontsize=8, color='black')

plt.show()
```

### Stage Analysis  
- What was the win pattern in qualifiers vs league stage?
- Did toss matter more in knockout games?

```python
#How many matches in each stage? 
matches_stage = df['stage'].value_counts()
print("Matches in each stage")
print(matches_stage)

ax = sns.barplot(x=matches_stage.index, y=matches_stage.values, palette='Set2', edgecolor='black')
for container in ax.containers: 
    ax.bar_label(container, fmt='%d', padding=4)

plt.title("Total Matches in each stage")
plt.xlabel("Stage")
plt.ylabel("Matches")
plt.show()
```

```python
#Which stage had higher average first innings scores?
avg_stage = df.groupby('stage')['first_ings_score'].mean()
ax = sns.barplot(x=avg_stage.index, y=avg_stage.values, palette='hsv', edgecolor='black')
for container in ax.containers: 
    ax.bar_label(container, fmt='%.2f', padding=4)

plt.title("Average First Innings Score in each stage")
plt.xlabel("Stage")
plt.ylabel("Avg. Score")
plt.show()
```

```python
#Does high first innings score guarantee a win? 
df['batting_first_win'] = df['won_by'] == 'Runs'

batting_first_win_lose = df.groupby('batting_first_win')['first_ings_score'].mean()

lost_avg = batting_first_win_lose[False]
won_avg = batting_first_win_lose[True]
diff = won_avg - lost_avg
print(f"When batting first team WINS → avg score: {won_avg:.2f}")
print(f"When batting first team LOSES → avg score: {lost_avg:.2f}")
print(f"Difference: {diff:.2f} runs")


plt.figure(figsize=(7,5))
ax = sns.barplot(x=batting_first_win_lose.index, y=batting_first_win_lose.values, palette='hsv', edgecolor='black')
for container in ax.containers: 
    ax.bar_label(container, fmt='%.2f', padding=4)

ax.set_xticks([0, 1])
ax.set_xticklabels(['LOST', 'WON'])
plt.title("Average First Innings Score when batting first win vs lost")
plt.xlabel("WON vs LOST")
plt.ylabel("Avg. 1st ing Score")
plt.show()

```

```python
# Did toss matter more in knockout games?
toss_matter = df.groupby('stage')['toss_winner == match_winner'].value_counts(normalize=True)

yes_only = toss_matter.xs('Yes', level=1) * 100 

print(f"Toss impact in Group stage : {yes_only['Group']:.2f}%")
print(f"Toss impact in Playoff     : {yes_only['Playoff']:.2f}%")

ax = sns.barplot(x=yes_only.index, y=yes_only.values, palette='hsv', edgecolor='black')
for container in ax.containers: 
    ax.bar_label(container, fmt='%.2f', padding=4)

plt.axhline(50, color='red', linestyle='dashed', linewidth=1.5, label='50% baseline')
plt.title("Toss impact across stages")
plt.xlabel("Stage")
plt.ylabel("Percentage")
plt.legend()
plt.show()
```
